```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/drive/MyDrive/heart.csv")
df.head()
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['Not Desease','Desease'])
plt.title("Target Distribution")
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['red
','yellow'])
plt.legend(["Haven't Disease","Have Disease"])
plt.title('Heart Disease by Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='blue
',label='Not Disease')
sns.distplot(df.age[df['target']==1],bins=30,color='yellow',label='Disease')
plt.legend()
plt.title('Heart Disease distibution for Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
